```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
# from tqdm import tqdm
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
    """Preprocess the given text by tokenising it, removing any stop words, 
    replacing each remaining token with its lemma (base form), and discarding 
    all lemmas that contain non-alphabetical characters.

    Arguments:
      text (str): The text to preprocess.

    Returns:
      The list of remaining lemmas after preprocessing (represented as strings).
    """
    lemma = []
    # text.split(' ')
    doc = nlp(text)
    # print(text)
    for token in doc:
      if token.is_alpha == True and token.is_stop == False:
        # print(token)
        lemma.append(token.lemma_.lower())
        # print(lemma)
    return lemma
```

```python
# Automatically Generated CFs
auto_df = pd.read_csv('bert-base_512_selected.csv')
auto_cf = auto_df.drop(labels='original', axis=1)
auto_cf['label'] = auto_cf['label'].apply(lambda x: 1 if x == 0 else 0)

# Human Generated CFs Kaushik et.al (Train and Test) Combined
human_cf_train = pd.read_csv('https://raw.githubusercontent.com/acmi-lab/counterfactually-augmented-data/master/sentiment/combined/train.tsv', delimiter='\t')
human_cf_test = pd.read_csv('https://raw.githubusercontent.com/acmi-lab/counterfactually-augmented-data/master/sentiment/combined/test.tsv', delimiter='\t')
human_cf_train = human_cf_train.map(lambda s: s.lower() if type(s) == str else s)
human_cf_test = human_cf_test.map(lambda s: s.lower() if type(s) == str else s)

human_cf_train['Sentiment'] = human_cf_train['Sentiment'].replace('negative', 0)
human_cf_train['Sentiment'] = human_cf_train['Sentiment'].replace('positive', 1)

human_cf_test['Sentiment'] = human_cf_test['Sentiment'].replace('negative', 0)
human_cf_test['Sentiment'] = human_cf_test['Sentiment'].replace('positive', 1)

#Human Generated CFs Kaushik et.al
cf_train = pd.read_csv('https://raw.githubusercontent.com/acmi-lab/counterfactually-augmented-data/master/sentiment/new/train.tsv', delimiter='\t')
cf_test = pd.read_csv('https://raw.githubusercontent.com/acmi-lab/counterfactually-augmented-data/master/sentiment/new/test.tsv', delimiter='\t')
cf_train = cf_train.map(lambda s: s.lower() if type(s) == str else s)
cf_test = cf_test.map(lambda s: s.lower() if type(s) == str else s)

cf_train['Sentiment'] = cf_train['Sentiment'].replace('negative', 0)
cf_train['Sentiment'] = cf_train['Sentiment'].replace('positive', 1)

cf_test['Sentiment'] = cf_test['Sentiment'].replace('negative', 0)
cf_test['Sentiment'] = cf_test['Sentiment'].replace('positive', 1)

# Original IMDB Reviews (Train and Test) from Kaushik et.al
original_train_df = pd.read_csv('https://raw.githubusercontent.com/acmi-lab/counterfactually-augmented-data/master/sentiment/orig/train.tsv',delimiter='\t')
original_train_df = original_train_df.map(lambda s: s.lower() if type(s) == str else s)
original_train_df['Sentiment'] = original_train_df['Sentiment'].replace('negative', 0)
original_train_df['Sentiment'] = original_train_df['Sentiment'].replace('positive', 1)

original_test_df = pd.read_csv('https://raw.githubusercontent.com/acmi-lab/counterfactually-augmented-data/master/sentiment/orig/test.tsv',delimiter='\t')
original_test_df = original_test_df.map(lambda s: s.lower() if type(s) == str else s)
original_test_df['Sentiment'] = original_test_df['Sentiment'].replace('negative', 0)
original_test_df['Sentiment'] = original_test_df['Sentiment'].replace('positive', 1)

# Combining Original Train and Automatically generated CFs to create augmented dataset.
combine_orig_cf_x = pd.DataFrame()
combine_orig_cf_y = pd.DataFrame()

combine_orig_cf_x['combined_reviews'] = pd.concat([original_train_df['Text'],auto_cf['counterfact']],ignore_index=True)
combine_orig_cf_x = combine_orig_cf_x.map(lambda s: s.lower() if type(s) == str else s)
# orig_cf_y['combined_labels'] = pd.concat([train_y['labels'],aug_df['label']],ignore_index=True)
combine_orig_cf_y['combined_labels'] = pd.concat([original_train_df['Sentiment'],auto_cf['label']],ignore_index=True)

combined_df = pd.concat([combine_orig_cf_x['combined_reviews'],combine_orig_cf_y['combined_labels']], axis=1)
combined_train_df, combined_test_df = train_test_split(combined_df, test_size=0.15, random_state=42)
```

```python
def evaluate(dataloader_val):

    model.eval()
    
    loss_val_total = 0
    predictions, true_vals = [], []
    
    for batch in dataloader_val:
        
        batch = tuple(b.to(device) for b in batch)
        
        inputs = {'input_ids':      batch[0],
                  'attention_mask': batch[1],
                  'labels':         batch[2],
                 }

        with torch.no_grad():        
            outputs = model(**inputs)
            
        loss = outputs[0]
        logits = outputs[1]
        loss_val_total += loss.item()

        logits = logits.detach().cpu().numpy()
        label_ids = inputs['labels'].cpu().numpy()
        predictions.append(logits)
        true_vals.append(label_ids)
        # print(logits)
    loss_val_avg = loss_val_total/len(dataloader_val) 
    
    predictions = np.concatenate(predictions, axis=0)
    true_vals = np.concatenate(true_vals, axis=0)
            
    return loss_val_avg, predictions, true_vals
```

```python
from torch.nn import Dropout
def trainer(data_train, data_val, model):
    for epoch in tqdm(range(1, epochs+1)):
        model.train()
        
        loss_train_total = 0
        progress_bar = tqdm(data_train, desc = 'Epoch {:1d}'.format(epoch),
                        leave = False,
                        disable = False)
        for batch in progress_bar:
            model.zero_grad()
            batch = tuple(b.to(device) for b in batch)
            inputs = {
                'input_ids': batch[0],
                'attention_mask': batch[1],
                'labels': batch[2]
            }
            outputs = model(**inputs)
            
            loss = outputs[0]
            loss_train_total += loss.item()
            loss.backward()
            
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            
            optimizer.step()
            scheduler.step()
            progress_bar.set_postfix({'training loss': '{:.3f}'.format(loss.item()/len(batch))})
        
        torch.save(model.state_dict(), f'Models/RoBERTA_orig_epoch{epoch}.model')
        
        tqdm.write('\nEpoch{epoch}')
        loss_train_avg = loss_train_total/len(data_train)
        tqdm.write(f'Training Loss: {loss_train_avg}')
        
        val_loss, predictions, true_vals = evaluate(data_val)
        val_f1 = f1_score_func(predictions, true_vals)
        tqdm.write(f'Validation Loss: {val_loss}')
        tqdm.write(f'F1 Score(weighted): {val_f1}')
```

## Shapley Values using SHAP

```python
import shap
import scipy as sp
import matplotlib
```

```python
def f(model, tokenizer, x):
    tv = torch.tensor(
        [tokenizer.encode(v, padding="max_length", max_length=500, truncation=True)
         for v in x]
    ).to(device)
    outputs = model(tv)[0].detach().cpu().numpy()
    scores = (np.exp(outputs).T / np.exp(outputs).sum(-1)).T
    val = sp.special.logit(scores[:, 1])  # use one vs rest logit units
    return val
```

```python
def load_model(file_path):
    model.load_state_dict(torch.load(file_path, map_location=torch.device(device)))
    model.eval()
    return model
```

### Shapley Value Force Plot for the SVM Classifier

```python
def svm_explainer(pipe_arg, train_data, test_data):
    training_instance = shap.sample(train_data, 50)
    explainer_svm = shap.KernelExplainer(pipe_arg.named_steps['SVC'].predict_proba, pipe_arg.named_steps['vectorizer'].transform(training_instance))
    shap_values_svm = explainer_svm.shap_values(pipe_arg.named_steps['vectorizer'].transform([test_data]))
    feature_names = pipe_arg.named_steps['vectorizer'].get_feature_names_out()
    return explainer_svm, shap_values_svm, feature_names
```

#### Shapley Value for Trained on Original Data and Tested on Original Data
